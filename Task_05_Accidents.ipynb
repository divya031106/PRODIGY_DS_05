```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("US_Accidents_March23.csv", nrows=50000)
df.head()
```

```python
df.columns
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
```

```python
df['Hour'].value_counts().sort_index().plot(kind='bar', figsize=(10,5))
plt.title("Accidents by Time of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
df['Weather_Condition'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Weather Conditions Causing Accidents")
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.show()
```

```python
df['State'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 States with Most Accidents")
plt.xlabel("State")
plt.ylabel("Accident Count")
plt.show()
```
